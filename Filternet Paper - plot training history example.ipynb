```python
###
# Plot for paper

figure(figsize=[6, 3], dpi=300)
ax = gca()
ddf = df.loc[:, ["val_loss", "val_f1","stopping_metric", ]]
ddf["lr_ratio"] = 1.0 * 0.93 ** ddf.index
ddf2 = df.loc[:, ["should_checkpoint", "done"]].astype(float)
ddf.plot(ax=ax)
ddf2.should_checkpoint[ddf2.should_checkpoint.astype(bool)].plot(style=".", ax=ax)
ddf2.done[ddf2.done.astype(bool)].plot(style="x", ax=ax)
xlim([-1, df.index.max() + 2])
ylim([0, 1.6])
legend(
    [
        "Stopping metric",
        "Validation loss",
        "Validation F1",
        "Learning rate ratio $lr_i$/$lr_0$",
        "Checkpoint (bool)",
        "Done (bool)",
    ],
    ncol=2
)
# title("Example Training History")
ylabel("Metric")
xlabel("Epoch")
tight_layout()
plt.savefig('training_history.png')
plt.show()
pass
```
